```python
# Definicija furijeove traansformacije sa vezbi, vise verujem njima nego iz biblioteke

def fourierTransform(x, N):
    # Ova funkcija izračunava Furijeovu transformaciju u N ekvidistantnih tačaka
    if len(x) < N:
        x = np.append(x, np.zeros(N - len(x)))
    
    X = np.zeros(N, dtype = 'complex')
    n = np.arange(N)
    for k in range(N):
        Omega_k = 2*np.pi/N*k
        X[k] = np.sum(x*np.exp(-1j*Omega_k*n))
    
    return X
```

## 1.1

```python
# Signal s(t)
fs = 100
Ts = 1/fs

# Proizvoljno odredjeni broj N za odredjivanje furijeove transformacije
N = 1024
f=np.arange(N)*fs/N

# Analogni signal x(t) i deklaracija Ni i fi
f1 = 1
f2 = 3
f3 = 7

N1 = 32
N2 = 128
N3 = 1024

duration1 = N1*Ts
duration2 = N2*Ts
duration3 = N3*Ts

# Odabiranje u Ts na intervalima N1, N2, N3

t1 = np.arange(0, duration1, Ts)
t2 = np.arange(0, duration2, Ts)
t3 = np.arange(0, duration3, Ts)

x1 = np.cos(2*np.pi*f1*t1) + 0.5*np.cos(2*np.pi*f2*t1) + 3*np.cos(2*np.pi*f3*t1)
x2 = np.cos(2*np.pi*f1*t2) + 0.5*np.cos(2*np.pi*f2*t2) + 3*np.cos(2*np.pi*f3*t2)
x3 = np.cos(2*np.pi*f1*t3) + 0.5*np.cos(2*np.pi*f2*t3) + 3*np.cos(2*np.pi*f3*t3)

X1 = fourierTransform(x1, N)
X2 = fourierTransform(x2, N)
X3 = fourierTransform(x3, N)
```

```python
#Plotovanje fur transformacija

fig, ax = plt.subplots(3,1,figsize=(8,8))
plt.subplots_adjust(wspace=100, left = 0.20)
ax[0].plot(f, np.abs(X1))
ax[0].set(ylabel = 'X$_1$(j$w$)', title ='Amplitudske karakteristike')
ax[0].set_xlim(0,fs/5)

ax[1].plot(f, np.abs(X2))
ax[1].set(ylabel = 'X$_2$(j$w$)')
ax[1].set_xlim(0,fs/5)

ax[2].plot(f, np.abs(X3))
ax[2].set(ylabel = 'X$_3$(j$w$)')
ax[2].set(xlabel = 'w')
ax[2].set_xlim(0,fs/5)

for i in range(2):
    ax[i].tick_params(axis='x', which='both', labelbottom=False)
```

```python
#Koristimo zadatu funkciju, i trazimo na sprektru od 0 do N/2 jer nam se ponavlja signal posle
#Takodje zadati odgovor skaliramo sa fs/n, jer mi uzimamo max odmiraka tako da amoramo da ih skaliramo na frekvenciju

# lokalni max X1
X1p = np.abs(X1)[0:N//2]
peaksPos1 = signal.argrelextrema(X1p, np.greater) 
peaksPos1 = peaksPos1[0]

maxPeaksPos1 = np.argsort(-X1[peaksPos1]) 
maxPeakIndex1 = peaksPos1[maxPeaksPos1]

print(maxPeakIndex1[0:3]*fs/N)

# lokalni max X2
X2p = np.abs(X2)[0:N//2]
peaksPos2 = signal.argrelextrema(X2p, np.greater) 
peaksPos2 = peaksPos2[0]

maxPeaksPos2 = np.argsort(-X2[peaksPos2]) 
maxPeakIndex2 = peaksPos2[maxPeaksPos2]

print(maxPeakIndex2[0:3]*fs/N)

# lokalni max X3
X3p = np.abs(X3)[0:N//2]
peaksPos3 = signal.argrelextrema(X3p, np.greater) 
peaksPos3 = peaksPos3[0]

maxPeaksPos3 = np.argsort(-X3[peaksPos3]) 
maxPeakIndex3 = peaksPos3[maxPeaksPos3]

print(maxPeakIndex3[0:3]*fs/N)
```

Kod signala x3 imamo veliku rezoluciju cime mozemo jasno daa vidimo njene pikove. Kod signala x2 vidimo dovoljno dobro. Spektar odabiranog signala x1 je neupotrebljiv.

```python
#Definisanje Hammingovih prozorskih funckija na ormercimaa N1,N2,N3
w1 = windows.hamming(len(x1), sym=False)
w2 = windows.hamming(len(x2), sym=False)
w3 = windows.hamming(len(x3), sym=False)
#Mnozenje sa prozorskim funkcijama
x1pp = x1*w1
x2pp = x2*w2
x3pp = x3*w3

X1pp = fourierTransform(x1pp, N)
X2pp = fourierTransform(x2pp, N)
X3pp = fourierTransform(x3pp, N)
```

```python

fs = 11250
Ts = 1 / fs

f1, f2, f3 = 1000, 1300, 1320

#N0 je broj odbiraka za odabiranje signala x(t) a N je broj odaabiranja prilikom racunanja Fur transf
N0 = 100
N= 10

def izracunaj(N):
    duration = N0 * Ts
    t = np.arange(0, duration, Ts)
    x = (5 * np.cos(2 * np.pi * f1 * t) +
         1000 * np.cos(2 * np.pi * f2 * t) +
         10 * np.cos(2 * np.pi * f3 * t))
    X =fft(x, N)
    f = np.arange(N) * fs / N
    return t, x, f, X

t, x, f, X = izracunaj(N)

fig, ax = plt.subplots(4, 1, figsize=(8, 8))
plt.subplots_adjust(bottom=0.3, hspace=2)  

ax[0].plot(t, x, 'b-')
ax[0].set(ylabel=r'$x(t)$', title='Originalni signal')
ax[0].set(xlabel='t')

ax[1].plot(f, np.abs(X), 'r-')
ax[1].set(ylabel=r'$|X(j\omega)|$', title='Furijeova transformacija')
ax[1].set(xlabel='w')

ax[2].plot(f,np.real(abs(X)), 'r-')
ax[2].set(ylabel=r'$Re{|X(j\omega)|}$', title='Realni deo')
ax[2].set(xlabel='w')

ax[3].plot(f,np.imag(abs(X)), 'r-')
ax[3].set(ylabel=r'$Im{|X(j\omega)|}$', title='Imaginarni deo')
ax[3].set(xlabel='w')

ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03])
slider = Slider(ax_slider, 'N', 10, 1000, valinit=N0, valstep=1)

# Update function
def update(N):

    ax[0].cla()
    ax[1].cla()
    ax[2].cla()
    ax[3].cla()
    
    N = int(N)
    t, x, f, X = izracunaj(N)

    ax[0].plot(t,x)
    ax[0].set(ylabel=r'$x(t)$', title='Originalni signal')
    ax[0].set(xlabel='t')
    
    ax[1].plot(f,abs(X), 'r-')
    ax[1].set(ylabel=r'$|X(j\omega)|$', title='Furijeova transformacija')
    ax[1].set(xlabel='w')

    ax[2].plot(f,np.real(abs(X)), 'r-')
    ax[2].set(ylabel=r'${Re|X(j\omega)|}$', title='Realni deo')
    ax[2].set(xlabel='w')

    ax[3].plot(f,np.imag(abs(X)), 'r-')
    ax[3].set(ylabel=r'$Im{|X(j\omega)|}$', title='Imaginarni deo')
    ax[3].set(xlabel='w')

    fig.canvas.draw_idle()
    

slider.on_changed(update)

plt.show()
```

```python
def forceAspect(ax, aspect=1):
    im = ax.get_images()
    extent =  im[0].get_extent()
    ax.set_aspect(abs((extent[1]-extent[0])/(extent[3]-extent[2]))/aspect)
```

```python
def dosSpectrogram(x, fs, window, noverlap, nfft, fMaxShow): 
    
    Nx = len(x)                                                 
    n = np.arange(Nx)                                           
    t = n/fs                                                    
    one_sided = True                                            
    for i in x:                                                 
        if np.imag(i) != 0:
            one_sided = False                                
    
    S = [np.abs(fft(x[i:i+nfft]*window,nfft)) for i in range(0, Nx - nfft, nfft - noverlap)] 
    f = np.arange(nfft)*fs/nfft 
    
    if one_sided == False:
        S = fft.fftshift(S)
        f = (np.arange(nfft)-nfft//2)*fs/nfft                  
    
    S = np.array(S).T 
    
    fMaxIndex = nfft*fMaxShow//fs
    fMaxIndex = round(fMaxIndex)
    return f, t, S[:][:fMaxIndex]
```

```python
def fft_radix_2(x):

    x = np.array(x, dtype=complex)
    N = len(x)
    a = math.ceil(math.log2(N))
    x = np.append(x, np.zeros(2**a-N))
    N=len(x)
    if N <= 2:
        return fourierTransform(x,N)
    else:
        k = np.arange(N//2)
        W = np.exp(-2j*np.pi*k/N)
        evens = fft_radix_2(x[::2])
        odds = fft_radix_2(x[1::2])
        return np.concatenate([evens + (W * odds), evens - (W * odds)])

    
    return 0
```

```python
# Signal iz tacke 6
fs = 112500
Ts = 1/fs

f1=1000
f2=1300
f3=1320

N0 = 4096

duration = N0*Ts
fk = np.arange(N0)*fs/N0
t=np.arange(0,duration,Ts)
x = 5*np.cos(2*np.pi*f1*t) + 1000*np.cos(2*np.pi*f2*t) + 10*np.cos(2*np.pi*f3*t)
```

```python
import scipy.fft as fft
X1 = fourierTransform(x,N0)
X2 = fft_radix_2(x)

fig, ax = plt.subplots(2,1,figsize=(8,6))
plt.subplots_adjust(wspace=100, left = 0.20)
ax[0].plot(fk, np.abs(X1))
ax[0].set(ylabel = 'X$_1$(j$w$)', title ='FFT definisan u 1. zadatku')
ax[0].set_xlim(0,1500)

ax[1].plot(fk, np.abs(X2))
ax[1].set(ylabel = 'X$_2$(j$w$)', title ='Radix 2 FFT')
ax[1].set(xlabel = 'w')
ax[1].set_xlim(0,1500)
```
